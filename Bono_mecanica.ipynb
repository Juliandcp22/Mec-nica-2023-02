```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t= np.linspace(0,10,100)
omega = 2
```

**Ecuaciones parametricas**





```python
def angulo(t):
  return t
```

```python
def distancia_radial(t):
  return 0.005*np.exp(2*t)+0.005*np.exp(-2*t)
```

```python
r=distancia_radial(t)
phi1= angulo(t)
```

```python
plt.plot(t,r)
plt.xlabel("t")
plt.ylabel("r(t)")
plt.grid()
```

```python
plt.plot(t,phi1)
plt.xlabel("t")
plt.ylabel("phi(t)")
plt.grid()
```

**Ã“rbitas**

```python
def orbita(omega,phi):
  return (0.005)*np.exp(omega*phi)+(0.005)*np.exp(-omega*phi)
```

```python
phi2= np.linspace(0, 10*np.pi,1000)
omega1= 0.01
omega2= 0.3
omega3= 20
```

```python
orbita1= orbita(omega1,phi2)
orbita2= orbita(omega2,phi2)
orbita3= orbita(omega3,phi2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="polar")
ax.plot(phi2,orbita1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="polar")
ax.plot(phi2,orbita2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="polar")
ax.plot(phi2,orbita3)
```
